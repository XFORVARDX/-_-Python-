### Режимы 

Изображение mode— это строка, определяющая тип и глубину пикселя изображения. Каждый пиксель использует полный диапазон битовой глубины. Таким образом, 1-битный пиксель имеет диапазон от 0 до 1, 8-битный пиксель имеет диапазон от 0 до 255 и так далее. Текущая версия поддерживает следующие стандартные режимы:

1(1-битные пиксели, черно-белые, хранятся по одному пикселю на байт)

L(8-битные пиксели, черно-белые)

P(8-битные пиксели, сопоставленные с любым другим режимом с использованием цветовой палитры)

RGB(3x8-битных пикселей, истинный цвет)

RGBA(4x8-битных пикселей, истинный цвет с маской прозрачности)

CMYK(4x8-битных пикселей, цветоделение)

YCbCr(3x8-битные пиксели, формат цветного видео)

Обратите внимание, что это относится к стандарту JPEG, а не к стандарту ITU-R BT.2020.

LAB(3x8-битных пикселей, цветовое пространство L*a*b)

HSV(3x8-битных пикселей, оттенок, насыщенность, цветовое пространство Value)

I(32-битные целые пиксели со знаком)

F(32-битные пиксели с плавающей запятой)

**изменить размер:**

Image.resize(size, resample=None, box=None, reducing_gap=None)

Возвращает измененную копию этого изображения.

ПАРАМЕТРЫ :
**size** — запрошенный размер в пикселях в виде двух кортежей: (ширина, высота).

**resample** — дополнительный фильтр передискретизации. Это может быть один из PIL.Image.Resampling.**NEAREST**, PIL.Image.Resampling.**BOX** **Текст, выделенный полужирным шрифтом**, PIL.Image.Resampling.**BILINEAR**, PIL.Image.**BICUBIC**, PIL.Image.Resampling.**HAMMING** ( режим «1» или «P»). 

box — необязательный 4-кортеж с плавающей запятой, обеспечивающий масштабируемую область исходного изображения. Значения должны быть в пределах прямоугольника (0, 0, ширина, высота). Если опущено или None, используется весь источник.

**Reduce_gap** — примените оптимизацию, изменив размер изображения в два шага. Во-первых, уменьшение изображения в целое число раз с помощью reduce(). Во-вторых, изменение размера с помощью обычной передискретизации. Последний шаг изменяет размер не менее чем в reducing_gap раз. reducing_gap может быть None (первый шаг не выполняется) или должен быть больше 1,0. Чем больше reducing_gap, тем ближе результат к справедливой передискретизации. Чем меньше reducing_gap, тем быстрее изменение размера. При reducing_gap большем или равном 3,0 результат в большинстве случаев неотличим от справедливой повторной выборки. Значение по умолчанию — None (без оптимизации).

ВОЗВРАЩАЕТ :
Объект Image

Гистограмма:


Возвращает гистограмму изображения. Гистограмма возвращается в виде списка количества пикселей, по одному для каждого значения пикселя в исходном изображении. Счетчики сгруппированы в 256 бинов для каждой полосы, даже если изображение имеет более 8 бит на полосу. Если изображение содержит более одного канала, гистограммы для всех каналов объединяются (например, гистограмма для изображения «RGB» содержит 768 значений).

С помощью этого метода двухуровневое изображение (режим «1») обрабатывается как изображение в оттенках серого («L»).

Если маска предоставлена, метод возвращает гистограмму для тех частей изображения, где изображение маски не равно нулю. Изображение маски должно иметь тот же размер, что и изображение, и быть либо двухуровневым изображением (режим «1»), либо изображением в градациях серого («L»).

ПАРАМЕТРЫ :
маска — необязательная маска.

extrema — необязательный набор экстремумов, заданных вручную.

Особые фильтры:

PIL.ImageFilter.GaussianBlur(radius=2) - Размывает изображение с помощью последовательности расширенных прямоугольных фильтров, которые приближаются к ядру Гаусса

PIL.ImageFilter.BoxBlur(radius)  - Размывает изображение, устанавливая для каждого пикселя среднее значение пикселей в квадратной рамке

PIL.ImageFilter.Kernel(size, kernel, scale=None, offset=0) - сверточный:

    size — Размер ядра, указанный как (ширина, высота). В текущей версии это должно быть (3,3) или (5,5).

    kernel – последовательность, содержащая веса ядра.

     scale – Масштабный коэффициент. Если задано, результат для каждого пикселя делится на это значение. По умолчанию используется сумма весов ядра.

    offset — смещение. Если задано, это значение добавляется к результату после его деления на коэффициент масштабирования.



PIL.ImageFilter.MedianFilter(size=3) - Выбирает среднее значение пикселя в окне заданного размера.

PIL.ImageFilter.MinFilter(size=3) - Выбирает наименьшее значение пикселя в окне заданного размера.

PIL.ImageFilter.MaxFilter(size=3) -  Выбирает наибольшее значение пикселя в окне заданного размера.

PIL.ImageFilter.ModeFilter(size=3) - Выбирает наиболее часто встречающееся значение пикселя в поле заданного размера. 

```python
from PIL import ImageFilter

im1 = image.filter(ImageFilter.BLUR)

im2 = image.filter(ImageFilter.MinFilter(3))
im3 = image.filter(ImageFilter.MinFilter)  # same as MinFilter(3)
```

```python
plt.subplot(1,2,1)
plt.imshow(np.array(image)[300:450,250:400])
plt.subplot(1,2,2)
plt.imshow(np.array(im1)[300:450,250:400])
plt.show()
```

```python
plt.subplot(1,2,1)
plt.imshow(np.array(image)[300:450,250:400])
plt.subplot(1,2,2)
plt.imshow(np.array(im2)[300:450,250:400])
plt.show()
```

```python
plt.subplot(1,2,1)
plt.imshow(np.array(image)[300:450,250:400])
plt.subplot(1,2,2)
plt.imshow(np.array(im3)[300:450,250:400])
plt.show()
```

Обработка пикселей
point()Этот метод использует функцию или таблицу поиска для обработки пикселей в изображении (например, операции контрастирования).

Преобразование пикселей


```python
# multiply each pixel by 1.2
out = image.point(lambda i: i * 1.2)
```

```python
plt.subplot(1,2,1)
plt.imshow(image)
plt.subplot(1,2,2)
plt.imshow(out)
plt.show()
```

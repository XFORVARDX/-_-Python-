```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
```

```python
def plot_verticles2D(vertices, filename = None, color = 'r', title = ''):
    # Create a new plot
    fig = plt.figure()

    ax = fig.add_subplot(111)
    f=3
    alf=pi/6
    xlime=f/(np.tan(alf)*0.04)
    ax.set_xlim([-xlime,xlime])
    ax.set_ylim([-xlime,xlime])
    x = [v[0] for v in vertices]
    y = [v[1] for v in vertices]
    print(x,y)

    ax.scatter(x, y, c=color, marker='o')
    ax.set_xlabel('X')
    ax.set_ylabel('Y')

    # Show or save the plot
    if filename is None:

        plt.grid()

        plt.show()

    else:
        plt.savefig(filename)


def plot_verticles(vertices, isosurf = False, filename = None, color = 'r', title = ''):
    # Create a new plot
    fig = plt.figure()

    ax = fig.add_subplot(111, projection='3d')

    x = [v[0] for v in vertices]
    y = [v[1] for v in vertices]
    z = [v[2] for v in vertices] 
    print(x)   
    if isosurf:
        ax.plot_trisurf(x, y, z, linewidth=0.2, antialiased=True)
    else:
        ax.scatter(x, y, z, c=color, marker='o')    
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Z')
    # Show or save the plot
    if filename is None:
        plt.show()
    else:
        plt.savefig(filename)
```
